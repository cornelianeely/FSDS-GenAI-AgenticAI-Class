# Superset | Subset | Disjoint

```python
s5 = {1, 2,3,4,5,6,7,8,9}
s6 = {3,4,5,6,7,8}
s7 = {10,20,30,40}
```

```python
s5.issuperset(s6) # superset acts as the 'father'
```

```python
s6.issuperset(s5)
```

```python
s6.issubset(s5) #subset acts as the 'son'
```

```python
s7.isdisjoint(s5) # Disjoint means to have no relation, i.e. stranger
```

```python
s7.isdisjoint(s6)
```

```python
s5 = {1, 2,3,4,5,6,7,8,9}
s6 = {3,4,5,6,7,8}
s7 = {10,20,30,40}
```

```python
s5.isdisjoint(s6)
```

```python
a3 = {1,2,3,4,5,6,7,8,9}
b3 = {13, 14, 15,16,17,18}
c3 = {10,20,30,40}
```

```python
b3.issubset(a3)
```

```python
b3.isdisjoint(c3)
```

# Range
- range is a built in function that generates a sequence of numbers. 
- It can take one, two, or three arguments: start, stop, and step. 
- The default start is 0, the default step is 1, and the stop is required. 
- The range function returns an iterable object that can be converted to a list or used in a loop.
- Range always stops one number before the stop number

```python
range() # range does NOT have an empty value, it needs at least one argument to work.
```

```python
range(5) 
# range with one argument will start at 0 and end at 4.
# It will not include the end value
```

```python
print(list(range(5)))
# range is a type of object
#it is not a list but it can be converted to a list
```

```python
list(range(5))
```

```python
list(range(1,5))
```

```python
list(range (1,10,3))
```

```python
list(range(1,10,5))
```

```python
range(10)
```

```python
r= range(10)
r
```

```python
for i in r:
    print(i)
```

```python
range(1,10,3,4)
# range only takes 3 arguments, start, stop, step. 
# It cannot take 4 arguments.
```
