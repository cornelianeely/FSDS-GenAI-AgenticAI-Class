# Set Operation

```python
# Union

A = {1,2,3,4,5}
B = {4,5,6,7,8}
C = {8,9,10}
```

```python
A.union(B) # .union() will return values that appear in both set A AND setB, basicalyl adding the sets together and removing any duplicates
```

```python
A|B # The "|" operator is another way to perform a union operation on sets, which will return a new set containing all unique elements from both sets A and B. (NO DUPLICATES)
```

```python
A.union(B,C) 
```

```python
    """
    Updates the set calling the update() method with union of A, B & C.
    For below example, Set A will be updated with union of A,B & C.
    """
    A.update(B,C) #Update the set, adding elements from all others.
    A
```

```python
# Intersection

A = {1,2,3,4,5}
B = {4,5,6,7,8}
```

```python
A.intersection(B) #.intersection() will return values that appear in both set A AND set B, which is the common elements between the two sets.
```

```python
A & B
```

```python
A.intersection(B) Intersection of A and B
```

```python
    """
    Updates the set calling the instersection_update() method with the intersection of A and B
    For below example, Set A will be updated with the intersection of A & B
    """
    A.intersection_update(B) # Update the set, keeping only elements found in it and all others.
    A
```

```python
# Difference

A = {1,2,3,4,5}
B = {4,5,6,7,8}
```

```python
A.difference(B) # .difference() will return values that appear in set A but not in set B, which is the difference between the two sets.
```

```python
B.difference(A)
```

```python
A - B # The "-" operator is another way to perform a difference operation on sets, which will return a new set containing all elements that are in set A but not in set B.
```

```python
B - A
```

```python
"""
Updates the set calling the difference_update() method with the difference of A and B.
For below example, Set A will be updated with the difference of A & B.
"""
B.difference_update(A) #Update the set, removing elements found in others.
B
```

```python
# Symmetric Difference
A = {1,2,3,4,5}
B = {4,5,6,7,8}
```

```python
A.symmetric_difference(B) # .symmetric_difference() will return values that appear in either set A or set B but not in both, which is the symmetric difference between the two sets.
```

```python
A ^ B # The "^" operator is another way to perform a symmetric difference operation on sets, which will return a new set containing all elements that are in either set A or set B but not in both.
```

```python
"""Updates the set calling the symmetric_difference_update() method with the symmetric difference of A and B.
For below example, Set A will be updated with the symmetric difference of A and B.
"""
A.symmetric_difference_update(B) # Update the set, keeping only elements found in either set, but not in both.
A
```

# Subset, Superset, Disjoint

```python
A = {1,2,3,4,5,6,7,8,9}
B = {3,4,5,6,7,8}
C = {10,20,30,40}
```

```python
B.issubset(A) # .issubset() will return True if all elements of set B are also present in set A, which means that B is a subset of A.
```

```python
A.issuperset(B) # .issuperset() will return True if all elements of set B are present in set A, which means that A is a superset of B.
```

```python
C.isdisjoint(A) # .isdisjoint() will return True if sets C and A have no elements in common, which means that they are disjoint sets.   
```

```python
B.isdisjoint(A) # .isdisjoint() will return False if sets B and A have elements in common, which means that they are not disjoint sets. 
```

# All/Any 
- The all() method returns:
-True- if all keys of the dictionary are true
- Falise if any key of the dictionary is false

- The any() fuction returns True if any key of the dictionary is True. If not any() returns False.

```python
mydict1 = {'Name': 'Asif', 'ID': 12345, 'DOB': 1991, 'Job': 'Analyst'}
mydict1
```

```python
all(mydict1) # all() will return True if all keys in the dictionary are truthy (not empty or zero), otherwise it will return False. In this case, all keys in 'mydict1' are non-empty strings, so it will return True.
```
