```python
# Tuple Slicing
```

```python
mytuple = ('one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight')
```

```python
mytuple[0:3] # returns all items from 0 (included) to 3rd (excluded) index
```

```python
mytuple[2:5] # returns all items from 2(included) to 5(excluded) index
```

```python
mytuple[:3] # returns first three items, 3rd index is excluded
```

```python
mytuple[:2] # returns first two items, 2nd index is excluded
```

```python
mytuple[-3:] # returns all items from -3(included) to the end
```

```python
mytuple[-2:] # returns all items from -2(included) to the end
```

```python
mytuple[-1] # returns last item 
```

```python
mytuple[:] # returns whole tuple
```
